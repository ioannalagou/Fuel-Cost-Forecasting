# 1Ô∏è‚É£ Create DB

```python
import sqlite3

# Connect / create db
conn = sqlite3.connect("fleet01.db")
cursor = conn.cursor()

# drop table if it exists
cursor.execute("DROP TABLE IF EXISTS fuel_usage")
cursor.execute("DROP TABLE IF EXISTS vehicles")

# Create vehicles table
cursor.execute("""
CREATE TABLE vehicles (
    vehicle_id INTEGER PRIMARY KEY,
    type TEXT,
    plate TEXT,
    brand TEXT,
    region TEXT
)
""")

# Create fuel_usage table
cursor.execute("""
CREATE TABLE fuel_usage (
    record_id INTEGER PRIMARY KEY,
    vehicle_id INTEGER,
    month TEXT,
    fuel_cost REAL,
    km_driven REAL,
    fuel_type TEXT,
    FOREIGN KEY(vehicle_id) REFERENCES vehicles(vehicle_id)
)
""")

conn.commit()
conn.close()
```

# 3Ô∏è‚É£ Insert Fake Data - Vehicles

```python
import sqlite3
import random

conn = sqlite3.connect("fleet01.db")
cursor = conn.cursor()

# -----------------------------
# FAKE VEHICLES
vehicle_types = ["Car", "Van", "Truck"]
brands = ["Toyota", "Ford", "VW", "Volvo", "Renault", "Skoda"]
regions = ["Athens", "Thessaloniki", "Patras", "Heraklion"]

num_vehicles = 100
vehicles_data = []

for i in range(1, num_vehicles + 1):
    vehicles_data.append((
        i,
        random.choice(vehicle_types),
        f"XYZ-{1000+i}",
        random.choice(brands),
        random.choice(regions)
    ))

cursor.executemany(
    "INSERT INTO vehicles (vehicle_id, type, plate, brand, region) VALUES (?, ?, ?, ?, ?);",
    vehicles_data
)

# -----------------------------
# FAKE FUEL USAGE
months = ["2025-09", "2025-10", "2025-11", "2025-12", "2026-01"]
fuel_records = []
record_id = 1

for vehicle_id in range(1, num_vehicles + 1):
    for month in months:
        fuel_type = "Diesel" if random.random() > 0.5 else "Petrol"  # 50 - 50
        km = random.randint(700, 3000)
        fuel_cost = round(random.uniform(100, 500), 2)

        fuel_records.append((
            record_id,
            vehicle_id,
            month,
            fuel_cost,
            km,
            fuel_type
        ))
        record_id += 1

cursor.executemany(
    "INSERT INTO fuel_usage (record_id, vehicle_id, month, fuel_cost, km_driven, fuel_type) VALUES (?, ?, ?, ?, ?, ?);",
    fuel_records
)

conn.commit()



```

# Data Loading

```python
import sqlite3
import pandas as pd
from IPython.display import display

# Connect to SQLite database
conn = sqlite3.connect("fleet01.db")

# Load tables into DataFrames
vehicles_df = pd.read_sql("SELECT * FROM vehicles;", conn)
fuel_df = pd.read_sql("SELECT * FROM fuel_usage;", conn)

# Display both tables nicely
display(vehicles_df.head())
display(fuel_df.head())
```

# Data Cleaning
Convert month to datetime and round up fuel_cost to nearest 10 for realism.


```python
import numpy as np

# Convert month to datetime
fuel_df["month"] = pd.to_datetime(fuel_df["month"])

# Round up fuel_cost to the nearest 10
fuel_df["fuel_cost"] = np.ceil(fuel_df["fuel_cost"] / 10) * 10

# Quick preview
from IPython.display import display
display(vehicles_df.head())
display(fuel_df.head())
```

# Make Synthetic Fuel Cost More Realistic (Region-Based Pricing)

In this step, fuel cost is recalculated based on:
- km driven
- fuel price range per region (1.60‚Ç¨ ‚Äì 2.10‚Ç¨ / liter)
- basic fuel consumption assumptions per vehicle type
This makes the synthetic dataset closer to a real-world business scenario.


```python
import numpy as np

# Fuel price per region: (Petrol, Diesel)
fuel_prices = {
    "Athens": (1.70, 1.50),
    "Thessaloniki": (1.80, 1.60),
    "Patras": (1.90, 1.75),
    "Heraklion": (2.00, 1.90)
}

# Consumption per 100km by vehicle type (example realistic values)
consumption_per_100km = {
    "Car": {"Petrol": 7, "Diesel": 6},
    "Van": {"Petrol": 9, "Diesel": 8}
}

# Merge region and type into fuel_df (Œ±ŒΩ Œ¥ŒµŒΩ œÑŒø Œ≠œáŒµŒπœÇ ŒÆŒ¥Œ∑)
fuel_df = fuel_df.merge(vehicles_df[["vehicle_id", "region", "type"]], on="vehicle_id", how="left")

# Calculate fuel cost per row
def calc_fuel_cost(row):
    fuel_type = row["fuel_type"]
    region = row["region"]
    vehicle_type = row["type"]
    
    # Get fuel price for region and type
    if region in fuel_prices:
        price_per_liter = fuel_prices[region][0] if fuel_type=="Petrol" else fuel_prices[region][1]
    else:
        price_per_liter = 1.8 if fuel_type=="Petrol" else 1.6  # default
    
    # Get consumption per 100km
    consumption = consumption_per_100km.get(vehicle_type, {"Petrol":7,"Diesel":6}).get(fuel_type,7)
    
    # fuel cost = km_driven / 100 * consumption * price_per_liter
    cost = row["km_driven"] / 100 * consumption * price_per_liter
    
    # Round to nearest 10
    cost = np.ceil(cost / 10) * 10
    return cost

fuel_df["fuel_cost"] = fuel_df.apply(calc_fuel_cost, axis=1)

# Quick preview
from IPython.display import display
display(fuel_df.head())
```

## Feature Engineering ‚Äì Preparing Dataset for Modeling

We are creating new features and aggregations from the fuel usage and vehicle tables to prepare a dataset suitable for machine learning models. 
This includes metrics like total and average fuel cost per vehicle and total kilometers driven.

```python
import pandas as pd
import numpy as np

# 1Ô∏è‚É£ Total fuel cost per vehicle
vehicle_total_cost = fuel_df.groupby("vehicle_id")["fuel_cost"].sum().reset_index()
vehicle_total_cost.rename(columns={"fuel_cost": "total_fuel_cost"}, inplace=True)

# 2Ô∏è‚É£ Average fuel cost per vehicle
vehicle_avg_cost = fuel_df.groupby("vehicle_id")["fuel_cost"].mean().reset_index()
vehicle_avg_cost.rename(columns={"fuel_cost": "avg_fuel_cost"}, inplace=True)

# 3Ô∏è‚É£ Total kilometers driven per vehicle
vehicle_total_km = fuel_df.groupby("vehicle_id")["km_driven"].sum().reset_index()
vehicle_total_km.rename(columns={"km_driven": "total_km"}, inplace=True)

# 4Ô∏è‚É£ Merge all features into a single DataFrame
features_df = vehicle_total_cost.merge(vehicle_avg_cost, on="vehicle_id")
features_df = features_df.merge(vehicle_total_km, on="vehicle_id")

# Quick preview
features_df.head()
```

## Define Targets and Features for Modeling

```python
import pandas as pd

# üîπ Sort by vehicle and month
fuel_df_sorted = fuel_df.sort_values(["vehicle_id", "month"])

# üîπ Create target: next month's fuel cost
fuel_df_sorted["next_month_fuel_cost"] = fuel_df_sorted.groupby("vehicle_id")["fuel_cost"].shift(-1)

# üîπ Drop rows without a next month
fuel_df_model = fuel_df_sorted.dropna(subset=["next_month_fuel_cost"])

# üîπ Regression features (you can add avg, total, km, region etc.)
X_regression = fuel_df_model[["km_driven", "fuel_cost"]]  # add any other features you want
y_regression = fuel_df_model["next_month_fuel_cost"]

# üîπ Classification target: high-cost vehicle next month (>20% above mean)
fuel_threshold = y_regression.mean() * 1.2
fuel_df_model["high_cost_flag"] = (y_regression > fuel_threshold).astype(int)
y_classification = fuel_df_model["high_cost_flag"]
X_classification = X_regression.copy()

# üîπ Quick preview
X_regression.head(), y_regression.head(), y_classification.head()

```

## Train-Test Split


```python
from sklearn.model_selection import train_test_split

# Train-test split (same split for both tasks)
X_train, X_test, y_train_reg, y_test_reg = train_test_split(
    X_regression, y_regression, test_size=0.2, random_state=42
)

y_train_clf = y_classification.loc[X_train.index]
y_test_clf  = y_classification.loc[X_test.index]

# Quick check
X_train.shape, X_test.shape, y_train_reg.shape, y_test_reg.shape
```

## Regression Model 1 ‚Äì Linear Regression


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Train Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train_reg)

# Predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluation metrics
mae_lr = mean_absolute_error(y_test_reg, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test_reg, y_pred_lr))

mae_lr, rmse_lr
```

```python
comparison_df = pd.DataFrame({
    "Actual": y_test_reg.values,
    "Predicted": y_pred_lr
})

comparison_df.head(10)
```

## Second Regression Model: Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest Regressor
rf_reg_model = RandomForestRegressor(random_state=42, n_estimators=100)
rf_reg_model.fit(X_train, y_train_reg)

# Predictions
y_pred_rf = rf_reg_model.predict(X_test)

# Evaluation metrics
mae_rf = mean_absolute_error(y_test_reg, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test_reg, y_pred_rf))

mae_rf, rmse_rf
```

## Model Comparison: Linear Regression vs Random Forest
Compare performance metrics (MAE, RMSE) and preview predictions for next month.

```python
# Create a comparison table for test set
comparison_df = pd.DataFrame({
    "Actual": y_test_reg.values,
    "LinearRegression_Pred": y_pred_lr,
    "RandomForest_Pred": y_pred_rf
})

# Display first 10 rows
comparison_df.head(10)

# Print metrics
print("Linear Regression: MAE =", mae_lr, ", RMSE =", rmse_lr)
print("Random Forest Regression: MAE =", mae_rf, ", RMSE =", rmse_rf)
```
